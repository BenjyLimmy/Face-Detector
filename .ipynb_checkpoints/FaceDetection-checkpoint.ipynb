## 3. Partition Unaugmented Data

### 3.1 Split Data into Train Test and Val 

```python
import shutil
from sklearn.model_selection import train_test_split
```

```python
def split_images(data_folder='./data/images', train_folder='./data/train/images', test_folder='./data/test/images', val_folder='./data/val/images', test_size=0.15, val_size=0.15, random_seed=42):
    # Create train, test, and val directories if they don't exist
    for folder in [train_folder, test_folder, val_folder]:
        os.makedirs(folder, exist_ok=True)

    # Get the list of image filenames
    image_files = os.listdir(data_folder)

    # Split the images into train, test, and val sets
    train_images, test_val_images = train_test_split(image_files, test_size=test_size + val_size, random_state=random_seed)
    test_images, val_images = train_test_split(test_val_images, test_size=val_size/(test_size + val_size), random_state=random_seed)

    # Copy images to the respective folders
    for image in train_images:
        shutil.move(os.path.join(data_folder, image), os.path.join(train_folder, image))

    for image in test_images:
        shutil.move(os.path.join(data_folder, image), os.path.join(test_folder, image))

    for image in val_images:
        shutil.move(os.path.join(data_folder, image), os.path.join(val_folder, image))
```

## 5. Build and Run Augmentation Pipeline
### 5.1 Run Augmentation Pipeline

```python
#### TEST TO SEE OUTPUT OF IMAGES WITH NO BBOXES

# Again, choose random json and photo with no face in it
label_path = os.path.join('data', 'train', 'labels', 'a5e4ef10-b5f0-11ee-8494-38d57a70299a.json')

img = cv2.imread(os.path.join('data', 'train', 'images', 'a5e4ef10-b5f0-11ee-8494-38d57a70299a.jpg'))
# plt.imshow(img)
coords = [0, 0, 0.00001, 0.00001]
augmented = augmentor(image=img, bboxes = [coords], class_labels=['face'])
augmented['class_labels'], augmented['bboxes']
# plt.imshow(augmented)
```

```python
NUM_AUG_IMS = 60

for partition in ['train', 'test', 'val']:
    #for each image path in set
    for image in os.listdir(os.path.join('data', partition, 'images')):
        #read image path
        img = cv2.imread(os.path.join('data', partition, 'images', image))
        
        #default annotation if image label does not exist, x_max and y_max must be slightly higher else error
        coords = [0, 0, 0.00001, 0.00001]
        label_path = os.path.join('data', partition, 'labels', f'{image.split(".")[0]}.json') #get json path
        
        # load bboxes coords if json exists
        if os.path.exists(label_path):
            with open(label_path, 'r') as f:
                label = json.load(f)
                
            coords = flat(label['shapes'][0]['points'])
            coords = list(np.divide(coords, [640, 480, 640, 480]))
            
        #create 60 images per base image
        try:
            for x in range(NUM_AUG_IMS):
                # Augment image
                augmented = augmentor(image=img, bboxes = [coords], class_labels=['face'])
                # Write augmented image into separate folder
                cv2.imwrite(os.path.join('aug_data', partition, 'images', f'{image.split(".")[0]}.{x}.jpg'),
                            augmented['image'])
                #create annotations
                annotation = {}
                annotation['image'] = image
                
                #if json exists
                if os.path.exists(label_path):
                    #if no bounding boxes (augmented out)
                    if len(augmented['bboxes']) == 0:
                        annotation['bbox'] = [0, 0, 0, 0]
                        #label output
                        annotation['class'] = 0
                    else:
                        annotation['bbox'] = augmented['bboxes'][0]
                        annotation['class'] = 1
                        
                else: # if did not have bounding box in the first place in original image
                    annotation['bbox'] = [0, 0, 0, 0]
                    annotation['class'] = 0
                    
                #write json labels into separate folder
                with open(os.path.join('aug_data', partition, 'labels', f'{image.split(".")[0]}.{x}.json'), 'w') as f:
                    json.dump(annotation, f)
        
        except Exception as e:
            print(e)
                
```

### 5.2 Load Augmented Images to Tensorflow Dataset

```python
resize = lambda x: tf.image.resize(x, (120, 120))
normalize = lambda x: x/255


def tf_map_images(partition):
    # Select all images
    images = tf.data.Dataset.list_files(os.path.join('aug_data', partition, 'images', '*'), shuffle = False)
    images = images.map(load_image).map(resize).map(normalize)
    return images    
```

## 6. Prepare Labels
### 6.1 Build Label Loading Function

```python
def load_labels(label_path):
    with open(label_path.numpy(), 'r', encoding = 'utf-8') as f:
        label = json.load(f)
        
    return [label['class']], label['bbox']
```

### 6.2 Load Labels to Tensforflow Dataset

```python
def tf_map_labels(partition):
    # Select all labels
    labels = tf.data.Dataset.list_files(os.path.join('aug_data', partition, 'labels', '*'), shuffle = False)
    labels = labels.map(lambda x: tf.py_function(load_labels, [x], [tf.uint8, tf.float16]))
    return labels
```

## 8. Build and Train Deep Learning Model

- Model is made up of two parts:
    1. Determining if it is face - Classification (0/1)
    2. Coordinates of bounding boxes for face - Regression

- VGG16 Base architecture
- Final Prediciton Layers (Classification & Regression)
- To achieve this, use Keras functional API
    * 2 different loss functions and combine
    * Combine - **localization loss**

### 8.1 Import Layers and Base Network

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv2D, Dense, GlobalMaxPooling2D
from tensorflow.keras.applications import VGG16
```

### 8.2 Download VGG16

- Model is a very big neural network that has been pretrained and built for image classification
- Only need fine tuning
- Conv2d , MaxPooling2D (for condensing infromation)
- Plugging in bulletproof architecture into our model

```python
# Instantiate VGG16 Model
vgg = VGG16(include_top = False) #exclude final layers (which are classification models and replace with our final layers)
```

```python
vgg.summary()
```

### 8.3 Build Instance of Network

```python
def build_model():
    # Input: Image dimensions
    input_layer = Input(shape = (120, 120, 3))
    
    # Instantiate VGG16 model excluding final layers and extend it to our input layer
    vgg = VGG16(include_top = False)(input_layer)
    
    # Classification Model
    f1 = GlobalMaxPooling2D()(vgg)
    class1 = Dense(2048, activation = 'relu')(f1)
    class2 = Dense(1, activation = 'sigmoid')(class1)
    
    # Bounding Box (Regression) Model
    f2 = GlobalMaxPooling2D()(vgg)
    regress1 = Dense(2048, activation = 'relu')(f2)
    regress2 = Dense(4, activation = 'sigmoid')(regress1)
    
    facetracker = Model(inputs = input_layer, outputs = [class2, regress2])
    return facetracker
```

## 9. Define Losses and Optimizers

### 9.1 Define Optimizer and LR

```python
batches_per_epoch = len(train)
lr_decay = (1./0.75 -1) / batches_per_epoch
```

```python
opt = tf.keras.optimizers.Adam(learning_rate=0.0001, weight_decay = lr_decay)
```

### 9.2 Create Localization Loss and Classification Loss

```python
def localization_loss(y_true, yhat):
    # calculate difference between x_min y_min hat and true value
    delta_coord = tf.reduce_sum(tf.square(y_true[:, :2] - yhat[:, :2])) 
    
    #albumenations format [x_min, y_min, x_max, y_max]
    # true width: x_max - x_min
    w_true = y_true[:,2] - y_true[:,0]
    #true height: y_max - y_min
    h_true = y_true[:,3] - y_true[:,1]
    
    #predicted value
    w_pred = yhat[:,2] - yhat[:,0]
    h_pred = yhat[:,3] - yhat[:,1]
    
    #calculate difference between height and width
    delta_size = tf.reduce_sum(tf.square(w_true - w_pred) + tf.square(h_true-h_pred))
    
    # return error in starting coordinate and dimension loss for finetuning
    return delta_coord + delta_size
```

```python
classloss = tf.keras.losses.BinaryCrossentropy()
regressloss = localization_loss
```

## 10. Train Neural Network

### 10.1 Create Custom Model Class

```python
class FaceTracker(Model): # inheritance from Model class requires (init, compile, train_step, call)
    def __init__(self, facetracker, **kwargs):
        super().__init__(**kwargs)
        self.model = facetracker
        
    def compile(self, opt, class_loss, localization_loss, **kwargs):
        super().compile(**kwargs)
        self.closs = class_loss
        self.lloss = localization_loss
        self.opt = opt
        
    def train_step(self, batch, **kwargs):
        
        X, y = batch
        #back propagation
        with tf.GradientTape() as tape:
            classes, coords = self.model(X, training = True)
            #forward
            batch_classloss = self.closs(y[0], classes)
            batch_localizationloss = self.lloss(tf.cast(y[1], tf.float32), coords)
            
            total_loss = batch_localizationloss + 0.5*batch_classloss
            #backward
            grad = tape.gradient(total_loss, self.model.trainable_variables)
            
        opt.apply_gradients(zip(grad, self.model.trainable_variables))
        return {"total loss": total_loss, "class loss": batch_classloss, "regress loss": batch_localizationloss}
    
    def test_step(self, batch, **kwargs):
        
        X, y = batch
        
        classes, coords = self.model(X, training = False)

        batch_classloss = self.closs(y[0], classes)
        batch_localizationloss = self.lloss(tf.cast(y[1], tf.float32), coords)

        total_loss = batch_localizationloss + 0.5*batch_classloss
            
        return {"total loss": total_loss, "class loss": batch_classloss, "regress loss": batch_localizationloss}
    
    def call(self, X, **kwargs):
        return self.model(X, **kwargs)
```

```python
model = FaceTracker(facetracker)
```

```python
model.compile(opt, classloss, regressloss)
```

### 11.3 Real Time Detection

```python
cap = cv2.VideoCapture(0)
while cap.isOpened():
    _ , frame = cap.read()
    frame = frame[50:500, 50:500, :]
    
    rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    resized = tf.image.resize(rgb, (120, 120))
    
    yhat = facetracker.predict(np.expand_dims(resized/255, 0))
    sample_coords = yhat[1][0]
    
    if yhat[0] > 0.5:
        # Controls Main Rectangle
        cv2.rectangle(frame,
                      tuple(np.multiply(sample_coords[:2], [450, 450]).astype(int)),
                      tuple(np.multiply(sample_coords[2:], [450, 450]).astype(int)),
                      (255, 0, 0), 2)
        # Controls the label rectangle
        cv2.rectangle(frame,
                      tuple(np.add(np.multiply(sample_coords[:2], [450, 450]).astype(int),
                                 [0,-30])),
                      tuple(np.add(np.multiply(sample_coords[:2], [450, 450]).astype(int),
                                 [80,0])),
                      (255, 0, 0), -1)
        # Controls text rendered
        cv2.putText(frame, 'face', tuple(np.add(np.multiply(sample_coords[:2], [450,450]).astype(int), [0, -5])),
                   cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2, cv2.LINE_AA)
        
    cv2.imshow('FaceTracker', frame)
    
    # q to exit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
        
cap.release()
cv2.destroyAllWindows()
```
